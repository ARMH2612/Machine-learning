# Example 1 - A simple cost function

$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x +1
```

```python
#Make data:
x_1 = np.linspace(start=-3, stop =3, num=100)
```

```python
# Plot:
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(X)',fontsize=16)
plt.plot(x_1, f(x_1))
plt.show()
```

## Slope and derivatives:

```python
def df(x):
    return 2*x+1
```

```python
# Plot:
plt.figure(figsize=[15,5])

#1 chart: cost fx:
plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim([0,8])
plt.title('Cost function', fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(X)',fontsize=16)
plt.plot(x_1, f(x_1), color='blue', linewidth=3)

#2 chart: derivative:
plt.subplot(1,2,2)
plt.title('Slpoe of the cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(X)',fontsize=16)
plt.grid()
plt.xlim([-2,3])
plt.ylim([-3,6])
plt.plot(x_1,df(x_1), color='skyblue', linewidth=5 )

plt.show()
```

## Pyhton Loops & Gradient Descent

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]
for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))
    if(step_size<precision):
        print(n)
        break
    

print("Local minimum occurse at: ",new_x)
print('Slope or df(x) value at this oint is: ', df(new_x))
print('f(x) value or cost at this value is: ',f(new_x))
```

$$ g(x) = x^4 - 4x^2 + 5 $$

```python
# Make some date:
x_2 = np.linspace(-2,2,1000)

def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x
```

```python
# Plot:
plt.figure(figsize=[15,5])

#1 chart: cost fx:
plt.subplot(1,2,1)

plt.xlim([-2,2])
plt.ylim([0.5,5.5])
plt.title('Cost function', fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('g(X)',fontsize=16)
plt.plot(x_2, g(x_2), color='blue', linewidth=3)

#2 chart: derivative:
plt.subplot(1,2,2)
plt.title('Slpoe of the cost function',fontsize=17)
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(X)',fontsize=16)
plt.grid()
plt.xlim([-2,2])
plt.ylim([-6,8])
plt.plot(x_2,dg(x_2), color='skyblue', linewidth=5 )

plt.show()
```
